```python
from urllib import request
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup
```

```python
def get_html(url):
   _html = ""
   resp = requests.get(url)
   if resp.status_code == 200:
      _html = resp.text
   return _html
```

```python
feature_list = ['제목','사기유형','발생지역','발생시기','피해금액', 'URL']
def soup_contents(start_article, num_articles):
    fraud = pd.DataFrame(np.zeros(shape=(num_articles, len(feature_list))), columns = feature_list)
    for i in range(num_articles):
        base_url ='http://news.kotra.or.kr/user/globalBbs/kotranews/27/globalBbsDataView.do?setIdx=256&dataIdx={}'
        url = base_url.format(start_article+i)
        html = get_html(url)
        soup = BeautifulSoup(html, 'html.parser')
        title = soup.find('title').text
        all_font_tags = soup.find_all("font")
        list_of_inner_text = [x.text for x in all_font_tags]
        if len(list_of_inner_text) > 4: #게시물 번호가 이어지지 않는 경우, 에러를 방지하기 위해서 list 값의 개수 체크
            fraud.loc[i, '제목'] = title
            fraud.loc[i, '사기유형'] = list_of_inner_text[1]
            fraud.loc[i, '발생지역'] = list_of_inner_text[2]
            fraud.loc[i, '발생시기'] = list_of_inner_text[3]
            fraud.loc[i, '피해금액'] = list_of_inner_text[4]
            fraud.loc[i, 'URL'] = url
    return fraud
```

```python
fraud_fin = soup_contents(146346,124)
```

```python
fraud_fin.to_csv('fraud_fin.csv')
```
