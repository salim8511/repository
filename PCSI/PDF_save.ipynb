```python
import sys
import requests
from bs4 import BeautifulSoup
import pdfkit
import pandas as pd

def get_url(dataIdx):
    base_url = 'http://news.kotra.or.kr/user/globalAllBbs/kotranews/album/2/globalBbsDataAllView.do?dataIdx={}'
    url = base_url.format(dataIdx)
    return url

def get_html(url):
   _html = ""
   resp = requests.get(url)
   if resp.status_code == 200:
      _html = resp.text
   return _html

def get_content(url):
    html = get_html(url)
    html_parsed = BeautifulSoup(html, 'html.parser')
    content = html_parsed.find("div", {"class": "global-news-view-wrap bbsPrintArea"})
    return content

def get_str(content):
    head_tags= '<head><meta charset="utf-8"><link rel="stylesheet" href="reset.css"><link rel="stylesheet" href="kotra_common.css"><link rel="stylesheet" href="sub.css"></head>'
    full_content = BeautifulSoup(head_tags)
    full_content.append(content)
    text = str(full_content)
    return text

def file_write(text):
    f = open("empty.html", "w")
    f.write(text)
    file = "empty.html"
    f.close()

def get_name(url):
    html = get_html(url)
    html_parsed = BeautifulSoup(html, 'html.parser')
    name = html_parsed.find("h5", {"class": "tit"}).text
    name = str(name)+".pdf"
    return name


def pdfdownload(file, name):
    pdfkit.from_file(file, name)
```

```python

data = pd.read_csv("dataidx.csv")
dataIdx = data["dataIdx"]
no = data["No"]
```

```python
for i in range(543,len(dataIdx)):
    url = get_url(dataIdx[i])
    content = get_content(url)
    name = get_name(url)
    number = no[i]
    name_new = str(number)+"."+str(name)
    text = get_str(content)
    file_write(text)
    pdfdownload("empty.html", name_new)
```
